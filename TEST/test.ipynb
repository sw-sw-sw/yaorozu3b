```python
import matplotlib.pyplot as plt
import numpy as np 
# データ
methods = [
    'calculate_distances',
    '_separation',
    '_cohesion',
    'predator_prey_forces',
    'limit_magnitude',
    'calculate_center_distances',
    'environment_forces',
    'calculate_forces'
]

times = [
    0.018058,
    0.034386,
    0.033845,
    0.132640,
    0.020970,
    0.018452,
    0.061304,
    0.008560
]

# 合計時間を計算
total_time = sum(times)

# パーセンテージを計算
percentages = [(time / total_time) * 100 for time in times]

# ラベルを作成（メソッド名と割合）
labels = [f'{method}\n({percentage:.1f}%)' for method, percentage in zip(methods, percentages)]

# 色を設定
colors = plt.cm.Set3(np.linspace(0, 1, len(methods)))

# 円グラフを作成
plt.figure(figsize=(12, 8))
plt.pie(times, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, labeldistance=1.1, pctdistance=0.8)
plt.title('TensorFlowSimulation Execution Time Distribution', fontsize=16)

# 凡例を追加
plt.legend(methods, title="Methods", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.axis('equal')  # 円を歪みのない円形に保つ
plt.tight_layout()  # レイアウトを調整

# グラフを表示
plt.show()
```
